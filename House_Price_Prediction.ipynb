```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import sklearn.datasets
```

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data,columns = house_price_dataset.feature_names)
```

```python
print(house_price_dataframe)
```

```python
house_price_dataframe['target'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.info()
```

```python
house_price_dataframe.describe()
```

```python
X = house_price_dataframe.drop(columns = 'target',axis=1)
Y = house_price_dataframe['target']
```

```python
print(X)
print(Y)
```

```python
correlation = house_price_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state =2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_pred = model.predict(X_train)
```

```python
r1_score = metrics.r2_score(Y_train,X_train_pred)
print(r1_score)
```

```python
r2_score = metrics.mean_absolute_error(Y_train,X_train_pred)
print(r2_score)
```

```python
plt.scatter(Y_train,X_train_pred)
plt.xlabel("Actual Prices")
plt.ylabel('Predicted Prices')
plt.title('Actual Prices VS Predicted Prices')
plt.show()
```

```python
X_test_pred = model.predict(X_test)
```

```python
r1_score = metrics.r2_score(Y_test,X_test_pred)
print(r1_score)
```

```python
r2_score = metrics.mean_absolute_error(Y_test,X_test_pred)
print(r2_score)
```

```python
plt.scatter(Y_test,X_test_pred)
plt.xlabel("Actual Prices")
plt.ylabel('Predicted Prices')
plt.title('Actual Prices VS Predicted Prices')
plt.show()
```
